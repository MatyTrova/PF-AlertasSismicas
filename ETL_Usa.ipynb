# **USA**

```python
import requests
import pandas as pd
import calendar
import io
import datetime
import calendar
```

```python
all_dfs = []

for i in range(1950, 2023):
    year = i
    starttime = f"{year}-01-01"
    endtime = f"{year}-12-31"
    url = f"https://earthquake.usgs.gov/fdsnws/event/1/query?format=csv&starttime={starttime}&endtime={endtime}&minmagnitude=4"
    response = requests.get(url)
    data = response.content
    df = pd.read_csv(io.StringIO(data.decode('utf-8')))
    all_dfs.append(df)   

# Datos actualizados
mes_actual = datetime.datetime.now().month
for mes in range (1, mes_actual +1):
    last_day = calendar.monthrange(2023, mes)[1]
    starttime = f"2023-{mes}-01"
    endtime = f"2023-{mes}-{last_day}"
    url = f"https://earthquake.usgs.gov/fdsnws/event/1/query?format=csv&starttime={starttime}&endtime={endtime}&minmagnitude=4"
    response = requests.get(url)
    data = response.content
    df = pd.read_csv(io.StringIO(data.decode('utf-8')))
    all_dfs.append(df)   


df_usa = pd.concat(all_dfs, ignore_index=True)
```

```python
df_usa["idCountry"] = 1
df_usa.head()
```

```python
df_usa.to_json('USA.json')
```
